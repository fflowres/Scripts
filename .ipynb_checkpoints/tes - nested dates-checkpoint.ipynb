```python
import datetime as dt
import pandas as pd

from CodernityDB.database import Database
from CodernityDB.database import RecordNotFound
from CodernityDB.database_thread_safe import ThreadSafeDatabase
dbName='dateDB'
#db=Database(dbName)
db=Database(dbName)
def spill():
    if(db.exists()):
        db.open()
        for curr in db.all('id'):
            print curr
        db.close()

def getTodayID(): ## normally from config file. maybe this is better?
    todayDate = str(getDateStartO())
    dayID = getDayRowID(todayDate)  
    return dayID
        

def getDayRow(rowID):
    dataop={}
    if(db.exists()):
        db.open()
        try:
            dataop = db.get('id', rowID, with_doc=True)
            db.close()
        except:
            return False
        else:
            return dataop
        
def getDayRowID(dateo): #gets row id by date
    if(db.exists()):
        db.open()
        for curr in db.all('id'):
            try:
                if curr['date'] == str(dateo):
                    dataop = curr['_id']
                    dataop = "".join(dataop)  #_id is returned as a list of charaters, must be concatenated to string
                    db.close()
                    return dataop #returns datestring
            except KeyError:
                continue
                
                #break
        #if it makes it here, entry doesnt exist
        db.close()
        return False #there is a probplem
    
def checkfordate(dateo): #checks for existance of that date in db
    if(db.exists()):
        db.open()
        for curr in db.all('id'):
            try:
                if curr['date'] == str(dateo):
                    db.close()
                    return True
            except KeyError:
                continue
                
                #break
        #if it makes it here, entry doesnt exist
        db.close()
        return False
    
def getDateStartO():
    d = dt.datetime.now()
    conv = str(d.year)+"-"+ str(d.month)+"-"+ str(d.day)+" "+"00:00:00.000000" #sets time to 0, therefore starts only at 0 in the morning
    #dto=dt.datetime.strptime(str(conv),"%Y-%m-%d %H:%M:%S.%f")
    return conv   

def makeDayRow():
    
    todayDate = str(getDateStartO())
    if (checkfordate(todayDate)==True): #already exists no need to write
        return False
    
    
    dayrow = {}
    dayrow["date"] = todayDate
    if(db.exists() == True):
        db.open()
        db.insert(dayrow)
        db.close()#must close first , no double opens
        dayID = getDayRowID(todayDate) # gets id of row just written
        
        ## would normally write to config file
        return dayID
    
def createDB():
    if(db.exists() == True):
        return False
    if(db.exists() == False):
        db.create()
        db.close()
        return True

        
def updateDB(id,data): #k.i.s.s.
    if(db.exists() == True):
        db.open()
        row = db.get('id', id, with_doc=True)
        row.update(data)
        db.update(row)
        db.close()
        return True
```

```python
def insertA(IP): #IP is the input questions
    ## must be linked to config file current day row
    nowRow = getTodayID()
    current = getDayRow(nowRow) ## getTodayID should be in config
    
    current.pop('_rev', None)
    current.pop('_id', None)
    ###
    accDate=str(dt.datetime.now()) # a more accureate date as to when it is written
    for question in IP.keys():
        try:
            current[question][accDate]=IP[question]
        except KeyError:
            print "KEY ERROR"
            current[question]={} # first time
            current[question][accDate]=IP[question]
            
        aggrK = current[question].keys()
        try:
            aggrK.remove("aggregate") #removes already summed from the summing process
        except ValueError:
            #doesnt exist yet, first insert
            pass
            
        aggregate=0
        for point in aggrK: # sum all key values (except sum)
            aggregate += current[question][point] #adds 
        current[question]['aggregate'] = aggregate #inserts /  overwrites existing sum
    updateDB(nowRow,current) # ID ,JSON / DICT
    return True


    
```

```python
def conf_dts_dto(dts):
    dto=datetime.datetime.strptime(dts,"%Y-%m-%d %H:%M:%S.%f")
    return dto

def getLastDate(dates):#input a list of dates
    dates.sort(reverse=True)
    return dates[0]


def DBtoDF():
    df = pd.DataFrame()
    if(db.exists()):
        db.open()
        
        for curr in db.all('id'): #get row
            tmp={} #holder that is constructed
            
            rowDate = curr["date"]
            
            questionsData = curr.keys() # returns a list
            questionsData.remove('_id')
            questionsData.remove('_rev')
            questionsData.remove('date')
            for question in questionsData: #get question from list
                
                if infoIndex[question]['multipoint']== "True": #display all points
                    multiP = curr[question].keys()
                    multiP.remove('aggregate')
                    tmp[question]={}
                    for point in multiP:
                        tmp[question][point] = curr[question][point]
                    
                    
                if infoIndex[question]['multipoint']== "False": #display only one
                    
                    datelist = curr[question].keys() #gets all dates within the question
                    datelist.remove('aggregate') #mixed in with the keys form that level (list)
                    
                    pointKey = getLastDate(datelist) #selects most recent date from list (keys)
                    tmp[question]={}
                    tmp[question][rowDate] = curr[question][pointKey] # replaces with single, most recent, point only
                    
                if infoIndex[question]['aggregate']== "True": #display only one aggregate in it's own column
                    

                    aggregate_key_name = str(question)+"_aggregate"
                    tmp[aggregate_key_name]={}
                    tmp[aggregate_key_name][rowDate] = curr[question]['aggregate'] # replaces with single
                    
                

            df = pd.concat([df,pd.DataFrame(tmp)])
        
        
        
        db.close()
    return df
```
