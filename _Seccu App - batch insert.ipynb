```python
import time
import datetime


class BatchInsert:
    def __init__(self,multi,multihours,dependantDataset,independantDataset):
        
        self.key = ''
        self.dbname = '_autopopulate'
        self.question_set = {'dependant_question':{'active':'True','typ':'slider','range':'0-100','aggregate':'False', 'multipoint':'True'},'independant_question':{'active':'True','typ':'slider','range':'0-100','aggregate':'False', 'multipoint':'True'}}
        
        self.multi = multi
        self.multihours = multihours
        self.shift = 0
        self.indset = independantDataset
        self.depset = dependantDataset
        self.error = 0
        
        if (len(self.indset) == len(self.depset)):
            self.ammount = len(self.depset)
            print 'ok'
        else:
            print 'dataset lengths not equal'
            self.error = 1
            
        if (self.error ==0):
            
            self.getfirstdate()
            self.timeshifts()
            self.shifts2utc()
            self.createdb()
            self.addquestionsdb()
            self.timewarp()
        
        
    def preparedb(self):
        self.createdb()
        self.addquestionsdb()
        return self
        
        
        
    def createdb(self):
        #create database
        from SecApp.DBBaseClass import DBSubsystem
        initdb = DBSubsystem(self.key,self.dbname).createDB()
        return self
        
        
    def addquestionsdb(self):
        #input questions
        
        from SecApp.DBQuestionClass import SecuQ
        initQ = SecuQ(self.key)
        initQ.questionInsert(self.question_set,descriptor='inclusive')
        return self
    
    
    def getfirstdate(self):
        if (self.multi != 0):
            self.first_step = (self.ammount / self.multi) 
        else:
            self.first_step = (self.ammount) 
        
        import datetime
        timenow = datetime.datetime.now()
        self.firsttime = timenow + datetime.timedelta(days=-(self.first_step + 1)) #always a day behind
        return self
        
    def timeshifts(self):
        self.shift_list = []
        self.lastdate = self.firsttime
        self.startdate = self.firsttime
        for ind in range(self.first_step):
            self.lastdate = self.startdate
            if (self.multi != 0):
                for subind in range(self.multi):
                    self.lastdate = self.lastdate + datetime.timedelta(hours=self.multihours)
                    
                    self.shift_list.append(self.lastdate)
                    
            self.startdate = self.startdate + datetime.timedelta(days=1)
        return self.shift_list
                    
    def shifts2utc(self):
        self.utcshifts = []
        import calendar
        for date in self.shift_list:
            self.utcshifts.append(calendar.timegm(date.utctimetuple()))
        return self
            
    def mytime(self): return self.utcshifts[self.shift]
    
    def shifttime(self):
        time.time = self.mytime
        return self
    
    
    def constructdataentry(self,index):
        response = {}
        response['dependant_question'] = self.depset[index]
        response['independant_question'] = self.indset[index]
        return response 
    
    
    
    #Population starts here
    def timewarp(self):
        from SecApp.DBInputClass import SecuIn
        initDI = SecuIn(self.key)
        for number in range(len(self.utcshifts)):
            self.shift = number # funky global indexing for time warping function
            self.shifttime() # changes perceived time
            #print datetime.datetime.now()
            
            
            responsedata = self.constructdataentry(number)
            initDI.questionDataIN(str(responsedata))
        return self
            
            
        
        
        
```
