```python
import datetime as dt
import pandas as pd

from CodernityDB.database import Database
from CodernityDB.database import RecordNotFound
from CodernityDB.database_thread_safe import ThreadSafeDatabase
dbName='dateDB'
#db=Database(dbName)
db=Database(dbName)
def spill():
    if(db.exists()):
        db.open()
        for curr in db.all('id'):
            print curr
        db.close()

def getTodayID(): ## normally from config file. maybe this is better?
    todayDate = str(getDateStartO())
    dayID = getDayRowID(todayDate)  
    return dayID
        

def getDayRow(rowID):
    dataop={}
    if(db.exists()):
        db.open()
        try:
            dataop = db.get('id', rowID, with_doc=True)
            db.close()
        except:
            return False
        else:
            return dataop
        
def getDayRowID(dateo): #gets row id by date
    if(db.exists()):
        db.open()
        for curr in db.all('id'):
            try:
                if curr['date'] == str(dateo):
                    dataop = curr['_id']
                    dataop = "".join(dataop)  #_id is returned as a list of charaters, must be concatenated to string
                    db.close()
                    return dataop #returns datestring
            except KeyError:
                continue
                
                #break
        #if it makes it here, entry doesnt exist
        db.close()
        return False #there is a probplem
    
def checkfordate(dateo): #checks for existance of that date in db
    if(db.exists()):
        db.open()
        for curr in db.all('id'):
            try:
                if curr['date'] == str(dateo):
                    db.close()
                    return True
            except KeyError:
                continue
                
                #break
        #if it makes it here, entry doesnt exist
        db.close()
        return False
    
def getDateStartO():
    d = dt.datetime.now()
    conv = str(d.year)+"-"+ str(d.month)+"-"+ str(d.day)+" "+"00:00:00.000000" #sets time to 0, therefore starts only at 0 in the morning
    #dto=dt.datetime.strptime(str(conv),"%Y-%m-%d %H:%M:%S.%f")
    return conv   

def makeDayRow():
    
    todayDate = str(getDateStartO())
    if (checkfordate(todayDate)==True): #already exists no need to write
        return False
    
    
    dayrow = {}
    dayrow["date"] = todayDate
    if(db.exists() == True):
        db.open()
        db.insert(dayrow)
        db.close()#must close first , no double opens
        dayID = getDayRowID(todayDate) # gets id of row just written
        
        ## would normally write to config file
        return dayID
    
def createDB():
    if(db.exists() == True):
        return False
    if(db.exists() == False):
        db.create()
        db.close()
        return True

        
def updateDB(id,data): #k.i.s.s.
    if(db.exists() == True):
        db.open()
        row = db.get('id', id, with_doc=True)
        row.update(data)
        db.update(row)
        db.close()
        return True
```
