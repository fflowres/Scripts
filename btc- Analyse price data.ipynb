```python
import math

def average(x):
    assert len(x) > 0
    return float(sum(x)) / len(x)

def pearson_def(x, y):
    assert len(x) == len(y)
    n = len(x)
    assert n > 0
    avg_x = average(x)
    avg_y = average(y)
    diffprod = 0
    xdiff2 = 0
    ydiff2 = 0
    for idx in range(n):
        xdiff = x[idx] - avg_x
        ydiff = y[idx] - avg_y
        diffprod += xdiff * ydiff
        xdiff2 += xdiff * xdiff
        ydiff2 += ydiff * ydiff

    return diffprod / math.sqrt(xdiff2 * ydiff2)
```

```python
def graphDB(key=None,value=None): #filter by
    jsonDat=[]
    if(db.exists()):
        db.open()
        for curr in db.all('id'):
            if key==None:
                jsonDat.append(curr)
            else:
                if curr[key]==value:
                    jsonDat.append(curr)
                else:
                    continue
        db.close()
    return jsonDat


def graphCAVir(useCAD=False):
    tmpDat=graphDB("exch","CAVirtex")
    tmpDF=pd.DataFrame(tmpDat)
    tmpDF=tmpDF.set_index('dt')
    tmpDF = tmpDF.drop('_id', 1)
    tmpDF = tmpDF.drop('_rev', 1)
    tmpDF = tmpDF.drop('kind', 1)
    if useCAD == False:
        tmpDF['high'] = tmpDF['high'] * exchangeRateCAD['rate']
        tmpDF['last'] = tmpDF['last'] * exchangeRateCAD['rate']
        tmpDF['low'] = tmpDF['low'] * exchangeRateCAD['rate']
        
    tmpDF['15_MA_last'] = pd.stats.moments.rolling_mean(tmpDF['last'], 15)
    return tmpDF

def graphBitSt():
    tmpDat=graphDB("exch","BitStamp")
    tmpDF=pd.DataFrame(tmpDat)
    tmpDF=tmpDF.set_index('dt')
    tmpDF = tmpDF.drop('_id', 1)
    tmpDF = tmpDF.drop('_rev', 1)
    tmpDF = tmpDF.drop('kind', 1)
    tmpDF = tmpDF.drop('timestamp', 1)
    tmpDF['15_MA_last'] = pd.stats.moments.rolling_mean(tmpDF['last'], 15)
    tmpDF['30_MA_last'] = pd.stats.moments.rolling_mean(tmpDF['last'], 30)
    
    return tmpDF

```

```python
#span: curve sample
#step: row steps
#phase: what row to start(reference at the end)
    
def autoSlice(dO,span,phase): 
    rowCount = len(dO.index) #get lst row
    pos = rowCount - phase
    start = pos - span
    end = pos
    
    return start,end
    
    
    
```

```python

def conv_dt(array):
    tmp = [dt.datetime.strptime(e, "%Y-%m-%d") for e in array]
    op = matplotlib.dates.date2num(tmp)
    return op

def conv_iso(array): # to num
    tmp = [dateutil.parser.parse(e) for e in array]
    op = matplotlib.dates.date2num(tmp)
    return op
```

```python
def pollbitstamp(last):
    #Scrape Data
    data=bitstampTicker()
    check=data['ask'],data['bid'],data['high'],data['last'],data['low'],data['timestamp'],data['volume'],data['vwap']
        
    #compare old vs new
    if (check != last): #if data is not repeted
        last=data['high'],data['volume'],data['last'],data['low']
            
        #write
        print "change written(BitStamp): " + str(data)
        insert("BitStamp","Ticker",data)
        return last #new values
    else:
        return last #repeated value
    
```

```python
def pollcavrtx(last_cavrtx):
    #Scrape Data
    cavrtx=cavirtexTicker()
    check=cavrtx['high'],cavrtx['volume'],cavrtx['last'],cavrtx['low']
        
    #compare old vs new
    if (check != last_cavrtx): #if data is not repeted
        last_cavrtx=cavrtx['high'],cavrtx['volume'],cavrtx['last'],cavrtx['low']
            
        #write
        print "change written(CAVirtex): " + str(cavrtx)
        insert("CAVirtex","Ticker",cavrtx)
        return last_cavrtx #new values
    else:
        return last_cavrtx #repeated value
    
```

```python
def pollchanges(interval=10): #poll interval in seconds
    #set initial nulls
    caVirtex=[]
    BitStamp=[]
    while(True): #repeat untill stop
        BitStamp=pollbitstamp(BitStamp)
        caVirtex=pollcavrtx(caVirtex) #loopback variable
        time.sleep(interval)
```
